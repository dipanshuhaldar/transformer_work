```python
# using pretrained models via pipeline
from transformers import pipeline
```

```python
# without specifying any specific model for text classification
pipe = pipeline('text-classification')
```

```python
pipe('this is great!')
```

```python
# specifying specific model for text classification
pipe = pipeline(model = 'roberta-large-mnli')
```
